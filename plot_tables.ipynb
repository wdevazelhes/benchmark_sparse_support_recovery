```python
import pandas as pd
import glob
import os

score_name = "objective_f1score"

# create a pandas dataframe from the results
parquet_files = glob.glob('./outputs/*.parquet')
latest_file = max(parquet_files, key=os.path.getmtime)
df = pd.read_parquet(latest_file)
df = df[['objective_name', 'solver_name', 'data_name', score_name]]
# df = 
# df.drop_duplicates(inplace=True)
# pivot the dataframe to show the score for each algorithm and dataset
pivot_df = df.pivot(index="solver_name", columns="data_name", values=score_name)

# print the pivot table
print(pivot_df)
```
